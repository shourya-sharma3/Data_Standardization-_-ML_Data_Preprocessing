# INTRODUCTION
The process of standardizing the data in a common format and range,
In process of standardisation, data becomes easy for machine to process and doesnt affect the true nature of data

Data standardization is a crucial step in machine learning preprocessing that ensures all features have a common scale, preventing certain variables from dominating due to their larger ranges. By transforming data to have a mean of 0 and a standard deviation of 1 using techniques like Z-score normalization, models such as logistic regression, KNN, and neural networks perform more efficiently. For example, if a dataset contains "Age" (ranging from 18-100) and "Salary" (ranging from 20,000-200,000), standardization ensures that both features contribute equally to the model’s learning process, leading to better accuracy and stability.

# **परिचय**  
डेटा को एक सामान्य प्रारूप और सीमा में मानकीकृत (स्टैंडर्डाइज़) करने की प्रक्रिया को **डेटा मानकीकरण** कहा जाता है। इस प्रक्रिया से डेटा मशीन द्वारा आसानी से संसाधित किया जा सकता है और यह डेटा के वास्तविक स्वरूप को प्रभावित नहीं करता।  

मशीन लर्निंग में डेटा मानकीकरण एक महत्वपूर्ण चरण है, जो यह सुनिश्चित करता है कि सभी फीचर्स समान पैमाने पर हों, ताकि किसी एक चर (वेरिएबल) का प्रभुत्व न हो, विशेष रूप से जब उनकी श्रेणियां (रेंज) भिन्न होती हैं। **ज़-स्कोर नॉर्मलाइज़ेशन** जैसी तकनीकों का उपयोग करके डेटा को इस तरह रूपांतरित किया जाता है कि इसका **औसत (Mean) 0** और **मानक विचलन (Standard Deviation) 1** हो। इससे लॉजिस्टिक रिग्रेशन, के-नियरस्ट नेबर्स (KNN), और न्यूरल नेटवर्क जैसे मॉडल अधिक कुशलता से काम करते हैं। उदाहरण के लिए, यदि किसी डेटासेट में "आयु" (18-100 के बीच) और "वेतन" (20,000-200,000 के बीच) जैसे फीचर्स हों, तो मानकीकरण यह सुनिश्चित करता है कि दोनों फीचर्स समान रूप से मॉडल की लर्निंग प्रक्रिया में योगदान दें, जिससे सटीकता और स्थिरता में सुधार होता है।

importing dependencies and importing dataset <br>

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('/content/boston.csv') #reading the dataset
```

```python
print(dataset)
```

```python
print(dataset.CRIM.std())
```

the above value is 8.601545105332487 which is not equal to or reletively closer to 1, our prefered choise is to get it closer to 1 if not equal to 1

the formula below is used for standardisation

उपर्युक्त मान **8.601545105332487** है, जो **1 के बराबर या उसके समीप नहीं है**।  
हमारा उद्देश्य इसे **संभवतः 1 के करीब लाना** है, यदि इसे पूरी तरह से **1 के बराबर नहीं किया जा सकता**।  

इस लक्ष्य को प्राप्त करने के लिए, नीचे दिया गया सूत्र **मानकीकरण (Standardization)** हेतु प्रयोग किया जाता है।

```python
scalar = StandardScaler()
scalar.fit(dataset)
StandardScaler(copy=True, with_mean=True, with_std=True)
x_standardised= scalar.transform(dataset)
print(x_standardised.std())
```
